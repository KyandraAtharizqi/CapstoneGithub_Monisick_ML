```python
import os
import shutil

# Define the base paths for both folders
frames_path = r"C:\Users\Rafi Kyandra\VSCODE\Semester 5\Bangkit MSIB\Capstone\nutrition5k\images\frames"
overhead_path = r"C:\Users\Rafi Kyandra\VSCODE\Semester 5\Bangkit MSIB\Capstone\nutrition5k\images\overhead"

# Initialize lists to track dish_id status
deleted_dishes = []
retained_dishes = []

# Iterate over each dish_id folder in the frames path
for folder in os.listdir(frames_path):
    folder_path = os.path.join(frames_path, folder)
    
    # Check if it's a directory
    if os.path.isdir(folder_path):
        overhead_folder_path = os.path.join(overhead_path, folder)
        
        # Check if the same dish_id exists in the overhead folder
        if os.path.exists(overhead_folder_path):
            retained_dishes.append(folder)  # Retain the dish_id
        else:
            # Delete the folder from frames if it doesn't exist in overhead
            try:
                shutil.rmtree(folder_path)  # Deletes the entire folder
                deleted_dishes.append(folder)  # Track deleted dish_id
                print(f"Deleted dish_id: {folder}")
            except Exception as e:
                print(f"Error deleting {folder}: {e}")

# List out the results
print("\nDeleted dish_ids:")
for dish in deleted_dishes:
    print(dish)

print("\nRetained dish_ids (existing in both folders):")
for dish in retained_dishes:
    print(dish)
```

```python
import os
import shutil

# Paths to the directories
base_path = r"C:\Users\Rafi Kyandra\VSCODE\Semester 5\Bangkit MSIB\CapstoneGithub\images"
overhead_path = os.path.join(base_path, "overhead")
frames_path = os.path.join(base_path, "frames")
theframes_path = r"C:\Users\Rafi Kyandra\VSCODE\Semester 5\Bangkit MSIB\Capstone\nutrition5k\images\theframes"

# Get list of dish IDs in each directory
overhead_dish_ids = set(os.listdir(overhead_path))
frames_dish_ids = set(os.listdir(frames_path))
theframes_dish_ids = set(os.listdir(theframes_path))

# Identify missing dish IDs in 'overhead' that are not in 'theframes'
missing_in_overhead = overhead_dish_ids - theframes_dish_ids

# Delete the unmatched folders from 'overhead'
print(f"Deleting {len(missing_in_overhead)} folders from 'overhead'...")

for dish_id in missing_in_overhead:
    folder_path = os.path.join(overhead_path, dish_id)
    if os.path.isdir(folder_path):
        shutil.rmtree(folder_path)
        print(f"Deleted: {dish_id}")

print("Deletion complete.")
```

```python
import os
import csv

# Paths
metadata_path = r"C:\Users\Rafi Kyandra\VSCODE\Semester 5\Bangkit MSIB\CapstoneGithub\metadata"
images_path = r"C:\Users\Rafi Kyandra\VSCODE\Semester 5\Bangkit MSIB\CapstoneGithub\images"
dish_csv_path = os.path.join(metadata_path, "dish.csv")
cleaned_csv_path = os.path.join(metadata_path, "csv_cleaned.csv")

# Get dish IDs from overhead and frames directories
overhead_dish_ids = {folder for folder in os.listdir(os.path.join(images_path, "overhead"))}
frames_dish_ids = {folder for folder in os.listdir(os.path.join(images_path, "frames"))}

# Combine valid dish IDs
valid_dish_ids = overhead_dish_ids.intersection(frames_dish_ids)

# Parse the dish.csv and create the cleaned file
matched_count = 0
discrepancy_ids = []

with open(dish_csv_path, "r") as infile, open(cleaned_csv_path, "w", newline="") as outfile:
    reader = csv.reader(infile)
    writer = csv.writer(outfile)
    
    for row in reader:
        dish_id = row[0]
        
        # Check if dish_id is valid
        if dish_id in valid_dish_ids:
            matched_count += 1
            # Write only the first six columns to the new file
            writer.writerow(row[:6])
        else:
            discrepancy_ids.append(dish_id)

# Print results
print(f"Total matched dish IDs: {matched_count}")
if discrepancy_ids:
    print(f"Discrepancy found for {len(discrepancy_ids)} dish IDs: {discrepancy_ids}")
else:
    print("No discrepancies found!")
```

```python
import os

# Define the base path to your frames directory
frames_base_dir = r"C:\Users\Rafi Kyandra\VSCODE\Semester 5\Bangkit MSIB\CapstoneGithub\images\frames"

# List all dish directories in the frames folder
dish_dirs = os.listdir(frames_base_dir)

# Initialize a list to store the number of images per dish
dish_image_counts = []

for dish_dir in dish_dirs:
    dish_path = os.path.join(frames_base_dir, dish_dir, "frames_sampled30")
    
    # Check if the directory exists
    if os.path.isdir(dish_path):
        # Get all the image files in the frames_sampled30 folder
        image_files = [f for f in os.listdir(dish_path) if f.endswith('.jpeg')]
        # Store the number of images
        dish_image_counts.append((dish_dir, len(image_files)))

# Sort the dishes by the number of images in descending order
dish_image_counts.sort(key=lambda x: x[1], reverse=True)

# Print the number of images for each dish, ranked from highest to lowest
for dish_id, count in dish_image_counts:
    print(f"{dish_id}: {count} images")
```
