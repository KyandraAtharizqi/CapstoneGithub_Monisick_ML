# Clean the CSV File

```python
import os
import pandas as pd

# File paths
input_csv = "metadata/dish.csv"
output_csv = "metadata/dishwithingredient_FINAL.csv"
frames_dir = r"C:\Users\Rafi Kyandra\VSCODE\Semester 5\Bangkit MSIB\CapstoneGithub\images\theframes"

# Read CSV with no header
data = pd.read_csv(input_csv, header=None)

# Assumed column structure
base_columns = ['dish_id', 'total_calories', 'total_mass', 'total_fat', 'total_carb', 'total_protein']
ingredient_columns = ['ingr_id', 'ingr_name', 'ingr_grams', 'ingr_calories', 'ingr_fat', 'ingr_carb', 'ingr_protein']

# Determine maximum number of ingredients in the CSV
ingredient_count = (data.shape[1] - len(base_columns)) // len(ingredient_columns)

# Dynamically create full column names for ingredients
columns = base_columns + [
    f"{field}_{i+1}" for i in range(ingredient_count) for field in ingredient_columns
]
data.columns = columns

# Extract ingredient names and combine them into a single column, excluding "deprecated" ingredients
def get_ingredient_names(row):
    ingredient_names = []
    for i in range(ingredient_count):
        ingr_name = row[f'ingr_name_{i+1}']
        if isinstance(ingr_name, str) and ingr_name.lower() != "deprecated":  # Exclude "deprecated" ingredients
            ingredient_names.append(ingr_name)
    return ', '.join(ingredient_names)

# Apply function to get ingredient names
data['ingredient_names'] = data.apply(get_ingredient_names, axis=1)

# Drop unnecessary ingredient columns
data = data[base_columns + ['ingredient_names']]

# Check if the dish_id folder exists in the frames directory
def folder_exists(dish_id):
    frames_path = os.path.join(frames_dir, str(dish_id))
    return os.path.isdir(frames_path)

# Apply the folder existence check
data['folder_exists'] = data['dish_id'].apply(folder_exists)

# Remove rows where folders do not exist
data_cleaned = data[data['folder_exists']].drop(columns=['folder_exists'])

# Export to CSV
data_cleaned.to_csv(output_csv, index=False)

# Print a confirmation message
print(f"New CSV file saved as {output_csv}")

```

# Augment Dishes with Less than 10 Images


```python
import os
import random
from PIL import Image, ImageEnhance

# Define the base path to your frames directory
frames_base_dir = r"C:\Users\Rafi Kyandra\VSCODE\Semester 5\Bangkit MSIB\CapstoneGithub\images\theframes"

# List all dish directories in the frames folder
dish_dirs = os.listdir(frames_base_dir)

# Function to apply random augmentations to an image
def augment_image(image):
    # Apply random brightness, contrast, and rotation
    enhancer = ImageEnhance.Brightness(image)
    image = enhancer.enhance(random.uniform(0.8, 1.2))  # Random brightness
    enhancer = ImageEnhance.Contrast(image)
    image = enhancer.enhance(random.uniform(0.8, 1.2))  # Random contrast
    image = image.rotate(random.uniform(-10, 10))  # Random rotation
    return image

# Loop through each dish directory
for dish_dir in dish_dirs:
    dish_path = os.path.join(frames_base_dir, dish_dir, "frames_sampled30")
    
    # Check if the directory exists
    if os.path.isdir(dish_path):
        # Get all the image files in the frames_sampled30 folder
        image_files = [f for f in os.listdir(dish_path) if f.endswith('.jpeg')]
        image_count = len(image_files)
        
        # If the number of images is less than 10, augment images
        if image_count < 10:
            print(f"Augmenting images for dish {dish_dir} (current: {image_count})")
            for i in range(10 - image_count):
                # Select a random image to augment
                random_image_path = os.path.join(dish_path, random.choice(image_files))
                with Image.open(random_image_path) as img:
                    # Apply augmentation
                    augmented_image = augment_image(img)
                    # Save the augmented image
                    new_image_name = f"augmented_{i + 1}.jpeg"
                    new_image_path = os.path.join(dish_path, new_image_name)
                    augmented_image.save(new_image_path)
```

# Split the dish into Training, Validation and Test Set

```python
import pandas as pd
import random

# File paths
input_csv = "metadata/dishwithingredient_FINAL.csv"
training_file = "training_dishes_nodeprecated_FINAL.txt"
validation_file = "val_dishes_nodeprecated_FINAL.txt"
testing_file = "test_dishes_nodeprecated_FINAL.txt"

# Load the CSV
data = pd.read_csv(input_csv)

# Get unique dish IDs
dish_ids = data['dish_id'].unique()

# Shuffle dish IDs to randomize the splits
random.shuffle(dish_ids)

# Split the data based on 92-4-4 ratio (4373 for training, 190 for validation, 190 for testing)
training_dishes = dish_ids[:4373]
validation_dishes = dish_ids[4373:4373 + 190]
testing_dishes = dish_ids[4373 + 190:4373 + 190 + 190]

# Ensure no overlap between splits
assert not set(training_dishes).intersection(set(validation_dishes)), "Overlap between training and validation!"
assert not set(training_dishes).intersection(set(testing_dishes)), "Overlap between training and testing!"
assert not set(validation_dishes).intersection(set(testing_dishes)), "Overlap between validation and testing!"

# Write splits to separate text files
with open(training_file, 'w') as f:
    f.write('\n'.join(map(str, training_dishes)))

with open(validation_file, 'w') as f:
    f.write('\n'.join(map(str, validation_dishes)))

with open(testing_file, 'w') as f:
    f.write('\n'.join(map(str, testing_dishes)))

# Print confirmation
print(f"Training dishes: {len(training_dishes)} written to {training_file}")
print(f"Validation dishes: {len(validation_dishes)} written to {validation_file}")
print(f"Testing dishes: {len(testing_dishes)} written to {testing_file}")
```
